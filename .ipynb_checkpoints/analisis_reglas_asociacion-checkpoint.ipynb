## 4. Preprocesamiento

Los algoritmos Apriori y FP-Growth de la librería mlxtend necesitan los datos en formato one-hot encoding: una tabla donde cada columna es un ítem y cada fila es una transacción. El valor es True si el ítem aparece en esa transacción, y False si no. El algoritmo ECLAT de pyECLAT acepta directamente listas de transacciones.

```python
lista_transacciones = []                                           

for fila in df['items']:                                           
    transaccion = fila.split(',')                                  
    lista_transacciones.append(transaccion)                        

print('Primeras 3 transacciones como listas:')                     
print(lista_transacciones[0])                                      
print(lista_transacciones[1])                                      
print(lista_transacciones[2])                                      
```

```python
from mlxtend.preprocessing import TransactionEncoder                             
                                                                                

encoder = TransactionEncoder()                                                   
arreglo_onehot = encoder.fit(lista_transacciones).transform(lista_transacciones) 
                                                                                 

df_onehot = pd.DataFrame(arreglo_onehot, columns=encoder.columns_)               
                                                                                 
print('Primeras 5 filas del one-hot encoding:')                                  
print(df_onehot.head())                                                          
print('Tamaño de la tabla:', df_onehot.shape)                                    
```

```python
columnas_del_dataset = list(df_onehot.columns)               
                                                              
                                                            
print('Ítems en el dataset:', sorted(items_distintos))       
```


## 5. Implementación de los algoritmos

### 5.1 Apriori

Imagina que trabajas para un supermercado, que tiene muchas sucursales en diferentes partes del país. Tienes muchas fuentes de información que pudieras utilizar para promocionar las ventas. Después de una revisión general, te das cuenta de que tienes una fuente de información valiosa, específicamente los tickets de venta. Ahí puedes obtener información valiosa. Para detectar oportunidades de venta puedes utilizar el agoritmo Apriori que revisa los productos que se compran con mayor frecuencia, y una vez identificados busca cuales productos se venden cuando se vende un producto principal. Primero busca pares, luego tríos y así sucesivamente. Sin embargo puedes encontrar productos con muy pocas ventas, y de cualquier manera el algoritmo busca pares, tríos, etc., que si parten de un producto que se vende poco, esas relaciones también producirán pocas ventas. El problema es que para identificar un producto de otros el algoritmo tiene que revisar todos los datos nuevamente. Por ello este algoritmo es lento. 

### 5.2 FP-Growth

En el caso de FP-Growth evita revisar muchas veces la base de datos completa, creando un árbol con los datos. Las revisiones acumulan datos en ramas comunes, de tal forma que el algoritmo sólo tiene que revisar la ramas para encontrar similitudes o patrones. Es como un índice de libro que te lleva sólo al capítulo que buscas, sin revisar todo el texto.

### 5.3 ECLAT

ECLAT cambia la forma de análisis no buscando el producto que se vende mucho y sus derivaciones, sino simplemente revisa cuantas veces aparece el producto principal en todos los tickets. Luego crea una lista de las transacciones en las que aparece dicho producto. Luego simplemente identifica cuales productos se compran junto con el producto identificado. Es rápido ya que su revisión es muy simple, pero con bases de datos muy grandes ocupa mucha memoria ram

```python
# Instalar
!pip install mlxtend pyECLAT

#Importar librerías
from mlxtend.frequent_patterns import apriori                                  
from mlxtend.frequent_patterns import fpgrowth                                  
from mlxtend.frequent_patterns import association_rules                         
from pyECLAT import ECLAT                                                       

print('Librerías importadas correctamente')                                     
```

```python
soporte_minimo = 0.05                          
                                              
confianza_minima = 0.3                        
print('Soporte mínimo:', soporte_minimo)      
print('Confianza mínima:', confianza_minima)  
```

### APRIORI

```python
tiempo_inicio_apriori = time.time()                                                   
                                                                                      

itemsets_apriori = apriori(df_onehot, min_support=soporte_minimo, use_colnames=True)  
                                                                                      
                                                                                      

tiempo_fin_apriori = time.time()                                                      
tiempo_apriori = tiempo_fin_apriori - tiempo_inicio_apriori                           

reglas_apriori = association_rules(itemsets_apriori, metric='confidence', min_threshold=confianza_minima) 
                                                                                         

print('Apriori terminó en', round(tiempo_apriori, 4), 'segundos')                        
print('Itemsets frecuentes encontrados:', len(itemsets_apriori))                         
print('Reglas generadas:', len(reglas_apriori))                                          
```

```python
print('Primeras 10 itemsets frecuentes (Apriori):')                                   
print(itemsets_apriori.head(10))                                                      
```

```python
print('Primeras 10 reglas de asociación (Apriori):')                                             # mensaje explicativo
print(reglas_apriori[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10))  # mostramos las 10 primeras reglas con 
                                                                                                 # las columnas más importantes
```

### FP GROWTH

```python
tiempo_inicio_fpgrowth = time.time()                                                    

itemsets_fpgrowth = fpgrowth(df_onehot, min_support=soporte_minimo, use_colnames=True)  

tiempo_fin_fpgrowth = time.time()                                                       
tiempo_fpgrowth = tiempo_fin_fpgrowth - tiempo_inicio_fpgrowth                          

reglas_fpgrowth = association_rules(itemsets_fpgrowth, metric='confidence', min_threshold=confianza_minima)  
                                                                                        

print('FP-Growth terminó en', round(tiempo_fpgrowth, 4), 'segundos')                    
print('Itemsets frecuentes encontrados:', len(itemsets_fpgrowth))                       
print('Reglas generadas:', len(reglas_fpgrowth))                                        
```

### ECLAT

```python
df_eclat_input = pd.DataFrame(lista_transacciones)           

print('Tipo de dato:', type(df_eclat_input))                 
                                                              
                                                             

tiempo_inicio_eclat = time.time()                            

objeto_eclat = ECLAT(data=df_eclat_input, verbose=True)      

indices_eclat, soporte_eclat = objeto_eclat.fit(             
    min_support=soporte_minimo,                              
    min_combination=1,                                       
    max_combination=4,                                       
    separator=' & ',                                         
    verbose=True                                             
)

tiempo_fin_eclat = time.time()                               
tiempo_eclat = tiempo_fin_eclat - tiempo_inicio_eclat        

print('ECLAT terminó en', round(tiempo_eclat, 4), 'segundos')  
print('Itemsets frecuentes encontrados:', len(soporte_eclat))  
```

```python
# Convertimos los resultados de ECLAT a un dataframe para verlos mejor

df_eclat_resultados = pd.DataFrame(list(soporte_eclat.items()), columns=['itemset', 'soporte'])  # creamos dataframe con los itemsets y 
                                                                                                 # su soporte
df_eclat_resultados = df_eclat_resultados.sort_values('soporte', ascending=False)                # ordenamos de mayor a menor soporte
print('Primeros 10 itemsets frecuentes (ECLAT):')                                                # mensaje explicativo
print(df_eclat_resultados.head(10))                                                              # mostramos los primeros 10
```

```python
# Generamos reglas de asociación manualmente a partir de los itemsets de ECLAT, ya que
# no tiene función directa de reglas como mlxtend, así que lo hacemos nosotros:


reglas_eclat_lista = []                                                         

for itemset_str, soporte_val in soporte_eclat.items():                          
    items_lista = itemset_str.split(' & ')                                      
    
    if len(items_lista) >= 2:                                                   
        for i in range(len(items_lista)):                                       
            consecuente = items_lista[i]                                        
            antecedente = [x for x in items_lista if x != consecuente]          
            
            antecedente_str = ' & '.join(antecedente)                           
            soporte_antecedente = soporte_eclat.get(antecedente_str, None)      
            
            if soporte_antecedente is not None and soporte_antecedente > 0:     
                confianza = soporte_val / soporte_antecedente                   
                if confianza >= confianza_minima:                               
                    reglas_eclat_lista.append({                                 
                        'antecedente': antecedente_str,                         
                        'consecuente': consecuente,                             
                        'soporte': soporte_val,                                 
                        'confianza': confianza                                  
                    })

reglas_eclat_df = pd.DataFrame(reglas_eclat_lista)                               
print('Reglas generadas por ECLAT:', len(reglas_eclat_df))                       
if len(reglas_eclat_df) > 0:                                                     
    print(reglas_eclat_df.head(10))                                              
```


## 6. Experimentos y resultados

### 6.1 Variación del soporte mínimo

Ejecutamos los tres algoritmos con diferentes valores de soporte mínimo: 1%, 2% y 5%.

```python
soportes_a_probar = [0.01, 0.02, 0.05]                                  

resultados_soporte = []                                                 

for soporte in soportes_a_probar:                                         
    print('Ejecutando con soporte mínimo:', soporte)                      
    
    # Apriori                                                             
    t0 = time.time()                                                      
    its_apr = apriori(df_onehot, min_support=soporte, use_colnames=True)  
    t_apr = time.time() - t0                                              
    try:                                                                  
                                                                          
        rls_apr = association_rules(its_apr, metric='confidence', min_threshold=confianza_minima) 
        n_rls_apr = len(rls_apr)                                          
    except:                                                               
        n_rls_apr = 0                                                     
    
    # FP Growth                                                           
    t0 = time.time()                                                      
    its_fp = fpgrowth(df_onehot, min_support=soporte, use_colnames=True)  
    t_fp = time.time() - t0                                               
    try:  # intentamos generar reglas
        rls_fp = association_rules(its_fp, metric='confidence', min_threshold=confianza_minima)  
        n_rls_fp = len(rls_fp)                                            
    except:                                                               
        n_rls_fp = 0                                                      
    
    # ECLAT                                                               
    t0 = time.time()                                                      
    obj_ec = ECLAT(data=df_eclat_input, verbose=False)                    
    _, sop_ec = obj_ec.fit(min_support=soporte, min_combination=1, max_combination=4, separator=' & ', verbose=False) 
    t_ec = time.time() - t0                                               
    
    # Guardamos resultados de este soporte:
    resultados_soporte.append({                                   
        'soporte': soporte,                                       
        'itemsets_apriori': len(its_apr),                         
        'itemsets_fpgrowth': len(its_fp),                         
        'itemsets_eclat': len(sop_ec),                            
        'reglas_apriori': n_rls_apr,                              
        'reglas_fpgrowth': n_rls_fp,                              
        'tiempo_apriori': round(t_apr, 4),                        
        'tiempo_fpgrowth': round(t_fp, 4),                        
        'tiempo_eclat': round(t_ec, 4)                            
    })

df_resultados_soporte = pd.DataFrame(resultados_soporte)          
print('\nTabla de resultados variando el soporte:')               
print(df_resultados_soporte.to_string(index=False))               
```

```python
# Gráfica: tiempo de ejecución vs soporte mínimo
fig, axes = plt.subplots(1, 3, figsize=(15, 4))                  

etiquetas_soporte = ['1%', '2%', '5%']                           

# Gráfica de tiempos
axes[0].plot(etiquetas_soporte, df_resultados_soporte['tiempo_apriori'], marker='o', label='Apriori', color='blue')    
                                                                                                                       
axes[0].plot(etiquetas_soporte, df_resultados_soporte['tiempo_fpgrowth'], marker='s', label='FP-Growth', color='green')
                                                                                                                       
axes[0].plot(etiquetas_soporte, df_resultados_soporte['tiempo_eclat'], marker='^', label='ECLAT', color='red')          
                                                                                                                       
axes[0].set_title('Tiempo de ejecución (segundos)')               
axes[0].set_xlabel('Soporte mínimo')                              
axes[0].set_ylabel('Segundos')                                    
axes[0].legend()                                                  
axes[0].grid(True)                                                

# Gráfica de itemsets frecuentes
axes[1].plot(etiquetas_soporte, df_resultados_soporte['itemsets_apriori'], marker='o', label='Apriori', color='blue')  
                                                                                                                       
axes[1].plot(etiquetas_soporte, df_resultados_soporte['itemsets_fpgrowth'], marker='s', label='FP-Growth', color='green')  
                                                                                                                  
axes[1].plot(etiquetas_soporte, df_resultados_soporte['itemsets_eclat'], marker='^', label='ECLAT', color='red')  
                                                                                                                  
axes[1].set_title('Número de itemsets frecuentes')                                               
axes[1].set_xlabel('Soporte mínimo')                                                             
axes[1].set_ylabel('Cantidad')                                                                   
axes[1].legend()                                                                                 
axes[1].grid(True)                                                                               

# Gráfica de reglas generadas
axes[2].plot(etiquetas_soporte, df_resultados_soporte['reglas_apriori'], marker='o', label='Apriori', color='blue')  
                                                                                                                     
axes[2].plot(etiquetas_soporte, df_resultados_soporte['reglas_fpgrowth'], marker='s', label='FP-Growth', color='green')  
                                                                                                                    
axes[2].set_title('Número de reglas de asociación')                                               
axes[2].set_xlabel('Soporte mínimo')                                                              
axes[2].set_ylabel('Cantidad')                                                                    
axes[2].legend()                                                                                  
axes[2].grid(True)                                                                                

plt.suptitle('Comparación de algoritmos variando el soporte mínimo', fontsize=14)                 
plt.tight_layout()                                                                                
plt.show()                                                                                        
```


## 7. Análisis y discusión

### 7.1 Interpretación de reglas representativas

```python
# Seleccionamos las reglas con mayor lift de Apriori para interpretarlas
reglas_apriori_ordenadas = reglas_apriori.sort_values('lift', ascending=False)        
                                                                                      
print('Top 5 reglas por lift (Apriori):')                                             
print(reglas_apriori_ordenadas[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(5).to_string(index=False))  
                                                                                      
```

```python
# Seleccionamos las reglas con mayor lift de FP-Growth para comparar
reglas_fpgrowth_ordenadas = reglas_fpgrowth.sort_values('lift', ascending=False)             
print('Top 5 reglas por lift (FP-Growth):')                                                  
print(reglas_fpgrowth_ordenadas[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(5).to_string(index=False))       
                                                                                            
```

```python
# Comparamos si Apriori y FP-Growth generan las mismas reglas
conjunto_apriori = set()                                                              
for fila in reglas_apriori.itertuples():  
    regla_str = str(sorted(fila.antecedents)) + '->' + str(sorted(fila.consequents))  
    conjunto_apriori.add(regla_str)                                                   

conjunto_fpgrowth = set()                                                             
for fila in reglas_fpgrowth.itertuples():                                             
    regla_str = str(sorted(fila.antecedents)) + '->' + str(sorted(fila.consequents))  
    conjunto_fpgrowth.add(regla_str)                                                  

reglas_solo_apriori = conjunto_apriori - conjunto_fpgrowth                           
reglas_solo_fpgrowth = conjunto_fpgrowth - conjunto_apriori                           
reglas_en_comun = conjunto_apriori & conjunto_fpgrowth                                

print('Reglas en común (Apriori y FP-Growth):', len(reglas_en_comun))                 
print('Reglas solo en Apriori:', len(reglas_solo_apriori))                            
print('Reglas solo en FP-Growth:', len(reglas_solo_fpgrowth))                         
print('(Ambos algoritmos deben encontrar exactamente las mismas reglas)')             
```

### 7.2 Comparación crítica entre algoritmos

**¿Por qué algunos algoritmos son más rápidos?**

- **Apriori** es el más lento porque genera **todos los posibles candidatos** antes de filtrar. Por ejemplo, si hay 19 ítems, genera hasta 2^19 = 524,288 posibles combinaciones. Aunque usa el principio de anti-monotonicidad para podar, sigue siendo costoso.

- **FP-Growth** es generalmente más rápido porque **comprime los datos en un árbol** y mina los patrones sin generar candidatos explícitamente. Hace solo 2 pasadas por los datos.

- **ECLAT** es eficiente cuando los datos son **densos** (muchos ítems por transacción) porque las intersecciones de listas son rápidas. Su desempeño puede variar con el dataset.

**¿Qué rol juegan las estructuras de datos internas?**

- Apriori usa **matrices horizontales** (cada transacción = lista de ítems).
- FP-Growth usa un **FP-tree** (árbol de prefijos comprimido).
- ECLAT usa **listas verticales** (cada ítem → lista de IDs de transacciones donde aparece).

**¿Cuándo deja de ser viable Apriori?**

Apriori se vuelve inviable cuando:
1. El dataset es muy grande (cientos de miles de transacciones).
2. Hay muchos ítems distintos (espacio de búsqueda exponencial).
3. El soporte mínimo es muy bajo (pocos ítems se filtran en cada pasada).

## 8. Conclusiones

En esta práctica se implementaron y compararon tres algoritmos de reglas de asociación sobre el dataset de canastas de mercado con 8,000 transacciones y 19 ítems distintos.

**Hallazgos principales:**

1. **Los tres algoritmos encuentran exactamente los mismos itemsets frecuentes** (y por lo tanto las mismas reglas de asociación) cuando se les da el mismo soporte mínimo. La diferencia está únicamente en la eficiencia, no en los resultados.

2. **FP-Growth fue consistentemente más rápido** que Apriori, especialmente con soportes mínimos bajos. Esto confirma la ventaja del FP-tree sobre la generación iterativa de candidatos.

3. **Al disminuir el soporte mínimo**, el número de itemsets frecuentes y reglas crece de forma significativa (sensibilidad alta). Esto implica que elegir el soporte correcto es crucial para obtener reglas útiles sin un número excesivo de ellas.

4. **La escalabilidad** de los tres algoritmos es razonable en datasets pequeños como este, pero FP-Growth y ECLAT escalan mejor que Apriori en datasets más grandes.

5. **Desde el punto de vista práctico**, reglas como `{detergente} → {suavizante}` o `{pan} → {mantequilla}` tienen sentido en un supermercado y podrían usarse para estrategias de marketing como la colocación de productos o descuentos cruzados.

6. **Apriori sigue siendo valioso** como algoritmo educativo por su simplicidad conceptual, aunque en producción se prefiere FP-Growth o algoritmos más modernos.
