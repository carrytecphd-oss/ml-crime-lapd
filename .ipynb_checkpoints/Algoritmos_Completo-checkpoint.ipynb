## Objetivo General

Analizar, implementar y comparar los algoritmos de aprendizaje por reglas de asociación **Apriori**, **FP-Growth** y **ECLAT** sobre un problema de análisis de canastas de mercado, con el fin de comprender sus diferencias conceptuales, computacionales y prácticas, evaluando su desempeño en términos de eficiencia, escalabilidad y calidad de los patrones descubiertos bajo distintos escenarios experimentales.

## 1. Instalación de librerías y carga de dependencias

```python
!pip install mlxtend pyECLAT

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import warnings

warnings.filterwarnings('ignore')

from mlxtend.frequent_patterns import fpgrowth, apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from pyECLAT import ECLAT

print("Librerías cargadas correctamente.")
```

---
## 2. Paso 1: Selección y exploración del dataset

Se trabaja con el dataset transaccional `dataset_market_basket_sintetico.csv`. Cada fila representa una transacción con un conjunto de ítems separados por coma. Se realiza un análisis exploratorio inicial para entender la estructura y distribución de los datos.

```python
# Carga del dataset
df = pd.read_csv('dataset_market_basket_sintetico.csv')

print(f"Dataset cargado correctamente → {len(df):,} transacciones")
print("\nPrimeras 10 transacciones:")
print(df.head(10))

# Número total de transacciones
print(f"\nNúmero total de transacciones: {len(df)}")

# Exploración de ítems
todos_los_items = []
for fila in df['items']:
    todos_los_items.extend(fila.split(','))

items_unicos = set(todos_los_items)
print(f"Número de productos distintos: {len(items_unicos)}")

# Estadísticas de longitud de canastas
longitudes = [len(fila.split(',')) for fila in df['items']]
print(f"Longitud promedio de una compra: {np.mean(longitudes):.2f} productos")
print(f"Compra más corta:  {min(longitudes)} productos")
print(f"Compra más larga:  {max(longitudes)} productos")

# Distribución de frecuencia de ítems
frecuencia = pd.Series(todos_los_items).value_counts()
print("\nLos 10 productos más comprados:")
print(frecuencia.head(10))
```

```python
# Visualización: distribución de los 15 ítems más frecuentes
plt.figure(figsize=(12, 5))
frecuencia.head(15).plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Distribución de frecuencia de los 15 ítems más populares', fontsize=13)
plt.xlabel('Producto')
plt.ylabel('Frecuencia absoluta')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

---
## 3. Paso 2: Preprocesamiento

Se convierten los datos al formato requerido por cada algoritmo:
- **Apriori y FP-Growth** (vía `mlxtend`): requieren una matriz one-hot encoding, donde cada columna es un ítem y cada fila es una transacción con valores `True/False`.
- **ECLAT** (vía `pyECLAT`): requiere un DataFrame donde cada fila es una transacción representada como una lista de ítems.

```python
# Convertir transacciones a lista de listas
transacciones = [fila.split(',') for fila in df['items']]

# One-hot encoding para Apriori y FP-Growth
te = TransactionEncoder()
te_ary = te.fit(transacciones).transform(transacciones)
df_onehot = pd.DataFrame(te_ary, columns=te.columns_)

print(f"DataFrame one-hot creado con forma: {df_onehot.shape}")
print(f"  → {df_onehot.shape[0]:,} transacciones  |  {df_onehot.shape[1]} ítems únicos")

# Formato para ECLAT
df_eclat = pd.DataFrame(transacciones)
print(f"\nDataFrame para ECLAT creado con forma: {df_eclat.shape}")
```

---
## 4. Paso 3: Implementación de los algoritmos

Se definen los parámetros comunes para asegurar una comparación justa entre los tres algoritmos.

```python
soporte_minimo  = 0.05   # 5%
confianza_minima = 0.60  # 60%

print(f"Soporte mínimo:   {soporte_minimo*100:.0f}%")
print(f"Confianza mínima: {confianza_minima*100:.0f}%")
```

### 4.1 Apriori

```python
print("Ejecutando Apriori...")
inicio = time.time()
itemsets_apriori = apriori(df_onehot, min_support=soporte_minimo, use_colnames=True)
tiempo_apriori = time.time() - inicio

reglas_apriori = association_rules(itemsets_apriori, metric="confidence", min_threshold=confianza_minima)

print(f"  Tiempo:              {tiempo_apriori:.4f} segundos")
print(f"  Itemsets frecuentes: {len(itemsets_apriori)}")
print(f"  Reglas generadas:    {len(reglas_apriori)}")
```

### 4.2 FP-Growth

```python
print("Ejecutando FP-Growth...")
inicio = time.time()
itemsets_fpgrowth = fpgrowth(df_onehot, min_support=soporte_minimo, use_colnames=True)
tiempo_fpgrowth = time.time() - inicio

reglas_fpgrowth = association_rules(itemsets_fpgrowth, metric="confidence", min_threshold=confianza_minima)

print(f"  Tiempo:              {tiempo_fpgrowth:.4f} segundos")
print(f"  Itemsets frecuentes: {len(itemsets_fpgrowth)}")
print(f"  Reglas generadas:    {len(reglas_fpgrowth)}")
```

### 4.3 ECLAT

```python
print("Ejecutando ECLAT...")
inicio = time.time()
eclat_model = ECLAT(data=df_eclat, verbose=False)
itemsets_eclat, soporte_eclat = eclat_model.fit(
    min_support=soporte_minimo,
    min_combination=1,
    max_combination=4,
    separator=' & ',
    verbose=False
)
tiempo_eclat = time.time() - inicio

# Generación manual de reglas para ECLAT
reglas_eclat_lista = []
for itemset_str, supp in soporte_eclat.items():
    items = itemset_str.split(' & ')
    if len(items) < 2:
        continue
    for i in range(len(items)):
        consecuente = items[i]
        antecedente = [x for x in items if x != consecuente]
        antecedente_str = ' & '.join(antecedente)
        supp_ante = soporte_eclat.get(antecedente_str)
        if supp_ante and supp_ante > 0:
            confianza = supp / supp_ante
            if confianza >= confianza_minima:
                reglas_eclat_lista.append({
                    'antecedente': antecedente_str,
                    'consecuente': consecuente,
                    'soporte':    supp,
                    'confianza':  confianza
                })

reglas_eclat_df = pd.DataFrame(reglas_eclat_lista)

print(f"  Tiempo:              {tiempo_eclat:.4f} segundos")
print(f"  Itemsets frecuentes: {len(soporte_eclat)}")
print(f"  Reglas generadas:    {len(reglas_eclat_df)}")
```

### 4.4 Resumen comparativo inicial

```python
comparacion = pd.DataFrame({
    'Algoritmo':           ['Apriori', 'FP-Growth', 'ECLAT'],
    'Tiempo (s)':          [round(tiempo_apriori, 4), round(tiempo_fpgrowth, 4), round(tiempo_eclat, 4)],
    'Itemsets frecuentes': [len(itemsets_apriori), len(itemsets_fpgrowth), len(soporte_eclat)],
    'Reglas generadas':    [len(reglas_apriori), len(reglas_fpgrowth), len(reglas_eclat_df)]
})

print("RESUMEN COMPARATIVO (soporte=5%, confianza=60%)")
print(comparacion.to_string(index=False))
```

```python
# Visualización de tiempos
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

algoritmos = ['Apriori', 'FP-Growth', 'ECLAT']
tiempos    = [tiempo_apriori, tiempo_fpgrowth, tiempo_eclat]
colores    = ['#1f77b4', '#2ca02c', '#d62728']

axes[0].bar(algoritmos, tiempos, color=colores, edgecolor='black')
axes[0].set_title('Tiempo de ejecución', fontsize=13)
axes[0].set_ylabel('Segundos')
axes[0].grid(axis='y', alpha=0.3)
for i, v in enumerate(tiempos):
    axes[0].text(i, v + max(tiempos)*0.01, f"{v:.4f}s", ha='center', fontsize=10)

reglas_count = [len(reglas_apriori), len(reglas_fpgrowth), len(reglas_eclat_df)]
axes[1].bar(algoritmos, reglas_count, color=colores, edgecolor='black')
axes[1].set_title('Reglas generadas', fontsize=13)
axes[1].set_ylabel('Número de reglas')
axes[1].grid(axis='y', alpha=0.3)
for i, v in enumerate(reglas_count):
    axes[1].text(i, v + max(reglas_count)*0.01, str(v), ha='center', fontsize=10)

plt.suptitle('Comparación inicial de algoritmos (soporte=5%, confianza=60%)', fontsize=14)
plt.tight_layout()
plt.show()
```

## 5. Paso 4: Diseño experimental

### 5.1 Experimento A — Variación del soporte mínimo

Se prueban tres valores de soporte (1%, 2%, 5%) con el dataset completo. Este experimento permite analizar la **sensibilidad al soporte mínimo** y la variación en el número de patrones descubiertos.

```python
soportes = [0.01, 0.02, 0.05]
resultados_soporte = []

for supp in soportes:
    print(f"\nProbando soporte mínimo = {supp*100:.0f}%")

    # Apriori
    t0 = time.time()
    its_a = apriori(df_onehot, min_support=supp, use_colnames=True)
    t_a = time.time() - t0
    rls_a = association_rules(its_a, metric="confidence", min_threshold=confianza_minima)

    # FP-Growth
    t0 = time.time()
    its_f = fpgrowth(df_onehot, min_support=supp, use_colnames=True)
    t_f = time.time() - t0
    rls_f = association_rules(its_f, metric="confidence", min_threshold=confianza_minima)

    # ECLAT
    t0 = time.time()
    e = ECLAT(data=df_eclat, verbose=False)
    _, sup_e = e.fit(min_support=supp, min_combination=1, max_combination=4,
                     separator=' & ', verbose=False)
    t_e = time.time() - t0

    resultados_soporte.append({
        'Soporte':           f"{supp*100:.0f}%",
        'Itemsets Apriori':  len(its_a),
        'Itemsets FP-Growth':len(its_f),
        'Itemsets ECLAT':    len(sup_e),
        'Reglas Apriori':    len(rls_a),
        'Reglas FP-Growth':  len(rls_f),
        'Tiempo Apriori':    round(t_a, 4),
        'Tiempo FP-Growth':  round(t_f, 4),
        'Tiempo ECLAT':      round(t_e, 4)
    })

df_soporte = pd.DataFrame(resultados_soporte)
print("\nResultados — Variación de soporte:")
print(df_soporte.to_string(index=False))
```

```python
# Visualización: tiempos vs soporte
x = df_soporte['Soporte']
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for col, color, label in [('Tiempo Apriori','#1f77b4','Apriori'),
                           ('Tiempo FP-Growth','#2ca02c','FP-Growth'),
                           ('Tiempo ECLAT','#d62728','ECLAT')]:
    axes[0].plot(x, df_soporte[col], marker='o', label=label, color=color)
axes[0].set_title('Tiempo de ejecución vs Soporte mínimo')
axes[0].set_xlabel('Soporte mínimo')
axes[0].set_ylabel('Segundos')
axes[0].legend()
axes[0].grid(alpha=0.3)

for col, color, label in [('Itemsets Apriori','#1f77b4','Apriori'),
                           ('Itemsets FP-Growth','#2ca02c','FP-Growth'),
                           ('Itemsets ECLAT','#d62728','ECLAT')]:
    axes[1].plot(x, df_soporte[col], marker='o', label=label, color=color)
axes[1].set_title('Itemsets frecuentes vs Soporte mínimo')
axes[1].set_xlabel('Soporte mínimo')
axes[1].set_ylabel('Número de itemsets')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.suptitle('Sensibilidad al soporte mínimo', fontsize=14)
plt.tight_layout()
plt.show()
```

### 5.2 Experimento B — Escalabilidad con distintos tamaños del dataset

Se prueban **3 submuestras crecientes** del dataset (30%, 60%, 100%) con soporte fijo al 5%. Este experimento evalúa cómo escala cada algoritmo al aumentar el volumen de datos.

> **Corrección respecto a la versión anterior:** se utilizan exactamente 3 submuestras (30%, 60%, 100%) conforme al enunciado, y se registran también las **reglas generadas** en cada iteración.

```python
# CORRECCIÓN: 3 submuestras (30%, 60%, 100%) + se registran reglas generadas
proporciones = [0.30, 0.60, 1.0]
resultados_tamano = []

for prop in proporciones:
    n = int(len(df_onehot) * prop)
    df_sub       = df_onehot.iloc[:n]
    trans_sub    = transacciones[:n]
    df_eclat_sub = pd.DataFrame(trans_sub)

    print(f"\nUsando {n:,} transacciones ({int(prop*100)}%)")

    # Apriori
    t0 = time.time()
    its_a = apriori(df_sub, min_support=0.05, use_colnames=True)
    t_a = time.time() - t0
    rls_a = association_rules(its_a, metric="confidence", min_threshold=confianza_minima)

    # FP-Growth
    t0 = time.time()
    its_f = fpgrowth(df_sub, min_support=0.05, use_colnames=True)
    t_f = time.time() - t0
    rls_f = association_rules(its_f, metric="confidence", min_threshold=confianza_minima)

    # ECLAT
    t0 = time.time()
    e = ECLAT(data=df_eclat_sub, verbose=False)
    _, sup_e = e.fit(min_support=0.05, min_combination=1, max_combination=4,
                     separator=' & ', verbose=False)
    t_e = time.time() - t0

    # Reglas ECLAT para este subconjunto
    rls_e_lista = []
    for itemset_str, supp in sup_e.items():
        items = itemset_str.split(' & ')
        if len(items) < 2:
            continue
        for i in range(len(items)):
            consecuente = items[i]
            antecedente_str = ' & '.join([x for x in items if x != consecuente])
            supp_ante = sup_e.get(antecedente_str)
            if supp_ante and supp_ante > 0:
                conf = supp / supp_ante
                if conf >= confianza_minima:
                    rls_e_lista.append(conf)

    resultados_tamano.append({
        'Tamaño':            f"{int(prop*100)}%",
        'Transacciones':     n,
        'Itemsets Apriori':  len(its_a),
        'Itemsets FP-Growth':len(its_f),
        'Itemsets ECLAT':    len(sup_e),
        'Reglas Apriori':    len(rls_a),
        'Reglas FP-Growth':  len(rls_f),
        'Reglas ECLAT':      len(rls_e_lista),
        'Tiempo Apriori':    round(t_a, 4),
        'Tiempo FP-Growth':  round(t_f, 4),
        'Tiempo ECLAT':      round(t_e, 4)
    })

df_tamano = pd.DataFrame(resultados_tamano)
print("\nResultados — Escalabilidad:")
print(df_tamano.to_string(index=False))
```

```python
# Visualización: escalabilidad
x2 = df_tamano['Tamaño']
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for col, color, label in [('Tiempo Apriori','#1f77b4','Apriori'),
                           ('Tiempo FP-Growth','#2ca02c','FP-Growth'),
                           ('Tiempo ECLAT','#d62728','ECLAT')]:
    axes[0].plot(x2, df_tamano[col], marker='o', label=label, color=color)
axes[0].set_title('Tiempo de ejecución vs Tamaño del dataset')
axes[0].set_xlabel('Tamaño del dataset')
axes[0].set_ylabel('Segundos')
axes[0].legend()
axes[0].grid(alpha=0.3)

for col, color, label in [('Reglas Apriori','#1f77b4','Apriori'),
                           ('Reglas FP-Growth','#2ca02c','FP-Growth'),
                           ('Reglas ECLAT','#d62728','ECLAT')]:
    axes[1].plot(x2, df_tamano[col], marker='o', label=label, color=color)
axes[1].set_title('Reglas generadas vs Tamaño del dataset')
axes[1].set_xlabel('Tamaño del dataset')
axes[1].set_ylabel('Número de reglas')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.suptitle('Escalabilidad de los algoritmos (soporte=5%, confianza=60%)', fontsize=14)
plt.tight_layout()
plt.show()
```

## 6. Paso 5: Análisis comparativo

### 6.1 Eficiencia computacional

Los resultados del resumen inicial y del experimento de soporte muestran diferencias notables en tiempo de ejecución:

- **Apriori** es el algoritmo más lento. Su ineficiencia radica en la generación iterativa de candidatos: en cada pasada genera todos los posibles itemsets de tamaño *k*, los contrastaa contra la base de datos completa y descarta los infrecuentes. Esto implica múltiples recorridos sobre el dataset y una explosión combinatoria de candidatos cuando el soporte mínimo es bajo.

- **FP-Growth** es significativamente más rápido. En lugar de generar candidatos, comprime las transacciones en una estructura de árbol llamada **FP-tree** (Frequent Pattern tree). El árbol almacena los ítems en orden de frecuencia y comparte prefijos entre transacciones, lo que reduce drásticamente el uso de memoria y el número de operaciones. La minería de patrones se realiza mediante la construcción recursiva de árboles condicionales sin necesidad de volver a leer el dataset original.

- **ECLAT** utiliza una representación **vertical** del dataset (tidlists): en lugar de almacenar transacciones como listas de ítems, almacena para cada ítem el conjunto de IDs de transacciones en que aparece. La intersección de estas listas permite calcular el soporte de itemsets sin recorrer el dataset fila por fila, lo que resulta muy eficiente en datasets densos o medianos.

### 6.2 Escalabilidad

Al aumentar el tamaño del dataset (30% → 60% → 100%):

- **Apriori** muestra el crecimiento más pronunciado en tiempo. Su complejidad es aproximadamente cuadrática o peor en función del número de ítems frecuentes, lo que lo hace poco viable en datasets grandes.
- **FP-Growth** escala de forma más controlada gracias a la compresión del FP-tree. Su crecimiento en tiempo es aproximadamente lineal al tamaño del dataset.
- **ECLAT** también escala bien, aunque en datasets muy grandes las tidlists pueden consumir mucha memoria RAM.

### 6.3 Sensibilidad al soporte mínimo

Al reducir el soporte del 5% al 1%:

- El número de itemsets frecuentes crece exponencialmente en los tres algoritmos, pero el impacto es especialmente grave para **Apriori**, cuyo tiempo de ejecución se dispara porque el espacio de candidatos se expande radicalmente.
- **FP-Growth** y **ECLAT** absorben mejor la reducción del soporte gracias a sus estructuras internas, aunque también incrementan su tiempo de cómputo.

### 6.4 ¿Cuándo deja de ser viable Apriori?

Apriori se vuelve impráctica cuando:
1. El soporte mínimo es muy bajo (< 2%), ya que el número de candidatos crece exponencialmente.
2. El dataset tiene muchas transacciones (> 50,000 filas) o ítems únicos (> 200), porque cada pasada requiere un recorrido completo.
3. Las transacciones son largas (muchos ítems por compra), lo que expande el espacio de candidatos de tamaño *k*.

En esos escenarios, **FP-Growth** es la alternativa recomendada por su eficiencia en memoria, y **ECLAT** cuando se dispone de suficiente RAM para almacenar las tidlists.

### 7.4 Discusión: relevancia, redundancia y comparación entre algoritmos

**Relevancia de las reglas:**  
Las reglas con mayor **lift** son las más relevantes desde el punto de vista estadístico, ya que indican que la co-ocurrencia entre antecedente y consecuente es significativamente mayor de lo que se esperaría por azar. Un lift > 1 confirma una asociación positiva real. Las reglas con soporte alto pero lift cercano a 1 pueden ser poco informativas (ítems que simplemente son populares en general).

**Redundancia:**  
Es esperable encontrar redundancia entre los tres algoritmos, ya que los tres buscan el mismo conjunto de itemsets frecuentes bajo los mismos parámetros. Por construcción, **Apriori y FP-Growth producen exactamente las mismas reglas** cuando se aplican sobre el mismo dataset y con los mismos umbrales; cualquier diferencia en el conteo es un artefacto de implementación. ECLAT puede presentar ligeras variaciones porque su generación de reglas es manual (no usa `association_rules` de mlxtend), pero los patrones de soporte subyacentes deben coincidir.

**Similitudes y diferencias entre algoritmos:**  
- Los tres producen el mismo conjunto de itemsets frecuentes (dado el mismo soporte mínimo), por lo que las reglas resultantes son equivalentes en contenido.
- La diferencia fundamental radica en la **estrategia de búsqueda** y las **estructuras de datos**: Apriori usa generación de candidatos en memoria principal (BFS), FP-Growth usa compresión en árbol (DFS implícito), y ECLAT usa intersección de listas de transacciones (DFS explícito).
- En términos prácticos, FP-Growth y ECLAT son preferibles a Apriori para cualquier dataset de tamaño no trivial.

**Relación con la base teórica:**  
- **Apriori** se basa en la propiedad antimonótona del soporte: si un itemset es infrecuente, todos sus superconjuntos también lo son. Esto permite podar el espacio de búsqueda, pero requiere múltiples pasadas por el dataset.
- **FP-Growth** explota la misma propiedad antimonótona pero evita la generación explícita de candidatos comprimiendo el dataset en un FP-tree, minando patrones mediante bases de patrones condicionales.
- **ECLAT** trabaja en el espacio vertical de tidlists: el soporte de un itemset se obtiene como el tamaño de la intersección de las tidlists de sus ítems componentes, lo que es computacionalmente eficiente y evita múltiples lecturas del dataset.

---
## 8. Conclusiones

1. Los tres algoritmos producen el mismo conjunto de itemsets frecuentes y, en consecuencia, reglas de asociación equivalentes bajo los mismos parámetros.
2. **FP-Growth** es el más eficiente en tiempo de ejecución para este dataset, seguido de cerca por **ECLAT**. **Apriori** es consistentemente el más lento.
3. La reducción del soporte mínimo impacta desproporcionadamente a Apriori, que experimenta un crecimiento exponencial en tiempo y número de candidatos.
4. En el experimento de escalabilidad, FP-Growth muestra el comportamiento más estable al crecer el dataset.
5. Desde una perspectiva práctica, **Apriori solo es recomendable para datasets pequeños o con soporte mínimo alto**. Para producción, FP-Growth o ECLAT son la elección adecuada.
